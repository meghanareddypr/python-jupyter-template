### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y = x * 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (Top-Left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = 2x')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Bar Plot (Top-Right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter Plot (Bottom-Left)
axs[1, 0].scatter(scatter_x, scatter_y, color='b', label='Data Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Pie Chart (Bottom-Right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
